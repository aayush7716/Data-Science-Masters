```python
data1 = pd.DataFrame({'key1':[1,2,3,4,6],
                      'key2':[4,5,6,7,8],
                      'key3':[2,3,4,5,6]
})
```

```python
data2 = pd.DataFrame({'key1':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]
})
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2)
```

```python
pd.merge(data1,data2,how = 'left')
```

```python
pd.merge(data1,data2,how = 'right')
```

```python
pd.merge(data1,data2,how = 'outer')
```

```python
pd.merge(data1,data2,how = 'cross')
```

```python
pd.merge(data1,data2,how = 'left', on = "key1")
```

```python
pd.merge(data1,data2,how = 'left', left_on='key2',right_on='key4')
```

```python
data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]},
                     index = ['a','b','c','d','e']
                    )
```

```python
data2 = pd.DataFrame({'key11':[1,2,4,5,6],
                      'key12':[4,5,6,7,8],
                      'key13':[3,4,5,6,6]},
                     index = ['a','b','h','i','j']
)
```

```python
data1
```

```python
data2
```

```python
data1.join(data2)
```

```python
data2.join(data1)
```

```python
data1.join(data2,how='inner')
```

```python
data1.join(data2,how='right')
```

```python
data1.join(data2,how='outer')
```

```python
data1.join(data2,how='cross')
```
